Scheint in dieser Stimme keine Ergebnisse zu geben. Dann reduzieren wir der Einfachkeithalber das Motiv vorübergehend auf `B- A`, da sollten wir Ergebnisse sehen.

```python
for i, thisNote in enumerate(thisPartNotes[:-3]):
    thisNoteName = thisNote.name
    if thisNoteName == 'B-':
        if thisPartNotes[i + 1].name == 'A':
            print('B-A found!')
```

Das sieht doch besser aus. Dann bauen wir damit unsere Funktionen und passen das Motiv dann später wieder an und durchsuchen den kompletten Corpus.

Nun bauen wir den Code noch um, dass wir das Motiv einfacher austauschen können.

```python
motive = ['B-', 'A']

for i, thisNote in enumerate(thisPartNotes[:-3]):
    print('---')
    motiveFound = True
    for j, thisMotiveNoteName in enumerate(motive):
        thisNoteName = thisPartNotes[i + j].name
        print(thisNoteName)
        if thisNoteName != thisMotiveNoteName:
            motiveFound = False
            break
    if motiveFound:
        print('found it!')
        

    
```

Dann schreiben wir das noch in eine Funktion und lassen uns die Indizes der Fundorte als Liste zurückgeben.

```python
motive = ['B-', 'A']#, 'B-']


def findMotive(thisPartNotes):
    results = []
    for i, thisNote in enumerate(thisPartNotes[:-len(motive) - 1]):
        motiveFound = True
        
        for j, thisMotiveNoteName in enumerate(motive):
            thisNoteName = thisPartNotes[i + j].name

            if thisNoteName != thisMotiveNoteName:
                motiveFound = False
                break
                
        if motiveFound:
            results.append(i)
            
    return results

print(findMotive(thisPartNotes))
```

Jetzt bauen wir noch eine Funktion, die uns die Noten im gewünschten Format zurückgibt.

Hinweis: damit wir später mit den Offsets der Noten nicht durcheinanderkommen, müssen wir noch `copy` importieren und beim Erstellen der Measures richtige Kopien anlegen.

```python
import copy
```

```python
def returnMotive(thisPart, firstNote):
    motiveMeasure = stream.Measure()
    
    # for i, _ in enumerate(motive):
    
    for i in range(len(motive)):
        thisNote = copy.deepcopy(thisPartNotes[firstNote + i])
        thisNote.duration.type = 'quarter'
        motiveMeasure.append(thisNote)
        
    motiveMeasure[0].lyric = str(firstNote)

    return motiveMeasure
```

Testen müssen wir sie natürlich auch.

Nun bauen wir noch eine Funktion drumrum, die die Lücke zwischen dem Corpus und dem Part schließt.

```python
def findMotiveInPiece(pieceMeta):
    piece = pieceMeta.parse()
    
    thisPieceMotives = stream.Stream()
    
    for thisPart in piece.parts:
        thisPartNotes = thisPart.recurse().getElementsByClass('Note')
        # thisPartnotes = thisPart.flat.notes

        thisPartMotivesFound = findMotive(thisPartNotes)
        for thisPartMotiveID in thisPartMotivesFound:
            thisMotivesMeasure = returnMotive(thisPartNotes, thisPartMotiveID)
        
            # Beschreibung
            thisMotiveFirstNoteID = int(thisMotivesMeasure[0].lyric)
            thisMotivePieceName = piece.metadata.title
            thisMotivePartName = thisPart.partName
            thisMotiveMeasure = thisPartNotes[thisMotiveFirstNoteID].measureNumber
            thisMotiveMeasureString = str(thisMotiveMeasure)
            
            thisMotivesMeasure[0].lyric = \
                  thisMotivePieceName + \
                  ', ' + thisMotivePartName + \
                  ', Measure: ' + thisMotiveMeasureString
        
            thisPieceMotives.append(thisMotivesMeasure)
        
    return thisPieceMotives
    
    
theMotiveResults = findMotiveInPiece(chorales[0])
theMotiveResults.show()
```

Jetzt haben wir das Motiv in dem einen Choral gefunden, nun wollen wir es im ganzen Corpus suchen.

```python
motive = ['B-', 'A', 'C', 'B']#, 'B-']
```

```python
motiveStream = stream.Stream()

for chorale in chorales:
    pieceMotivesMeasures = findMotiveInPiece(chorale)
    
    for measure in pieceMotivesMeasures:
        motiveStream.append(measure)
```

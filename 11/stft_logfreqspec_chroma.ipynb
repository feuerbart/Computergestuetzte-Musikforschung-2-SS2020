```
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa as lr
```

```
x, sr = lr.load('piano_major_scale.mp3')

stft = lr.stft(x)
plt.figure(figsize=(12, 4))
plt.imshow(10 * np.log10(np.abs(stft)), aspect='auto', origin='lower', extent=[0, np.shape(x)[0] / sr, 0, sr])
plt.colorbar()
plt.show()
```

```
x, sr = lr.load('piano_major_scale.mp3')

log_spec = lr.pseudo_cqt(x, sr=sr)
plt.figure(figsize=(12, 4))
plt.imshow(10 * np.log10(np.abs(log_spec)), aspect='auto', origin='lower', extent=[0, np.shape(x)[0] / sr, 0, np.shape(log_spec)[0]])
plt.colorbar()
plt.show()
```

```
x, sr = lr.load('piano_major_scale.mp3')

chroma = lr.feature.chroma_stft(x, sr=sr)
plt.figure(figsize=(12, 4))
plt.imshow(10 * np.log10(np.abs(chroma)), aspect='auto', origin='lower', extent=[0, np.shape(x)[0] / sr, 0, np.shape(chroma)[0]])
plt.colorbar()
plt.show()

print(np.shape(chroma))
```
